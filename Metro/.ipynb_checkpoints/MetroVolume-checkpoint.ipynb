```python
import os 
import gzip
from six.moves import urllib
import pandas as pd
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt 
import datetime
```

## Getting the data

```python
DOWNLOAD_ROOT = "http://archive.ics.uci.edu/ml/machine-learning-databases/00492/"
METRO_PATH = os.path.join("datasets" , "metro")
METRO_URL = DOWNLOAD_ROOT + "Metro_Interstate_Traffic_Volume.csv.gz"

def load_metro_data(metro_url = METRO_URL, metro_path = METRO_PATH):
    if not os.path.isdir(metro_path):
        os.makedirs(metro_path)
        
    gz_path = os.path.join(metro_path, "metro.csv.gz")
        
    urllib.request.urlretrieve(metro_url, gz_path)
        
    with gzip.open(gz_path) as f:
        data = pd.read_csv(f)

    return data
```

```python
metro = strat_train_set.drop("traffic_volume", axis=1)
metro_labels = strat_train_set["traffic_volume"].copy()
```

```python
#Now we have to fight with all of these categorical attributes I think there are 
#too many categories to use oneHotEncoding so Im going to use ordinalEncoding
#So lets create some functions that will deal with categorical attributes
from sklearn.base import BaseEstimator, TransformerMixin

class CategoricalDealer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self,X,y=None):
        return self
    def transform(self, X, y=None):
        
        holidays = X["holiday"].copy()
        for i in range(len(holidays)):
            holidays.iloc[i] = 0 if holidays.iloc[i] == "None" else 1
        dates = X["date_time"].copy()
        for i in range(len(dates)):
            dates.iloc[i] = (datetime.datetime.strptime(dates.iloc[i], '%Y-%m-%d %H:%M:%S').hour)/24
            
        return np.c_[holidays,dates]

#dealer = CategoricalDealer()
#metro_extra = dealer.transform(metro.values)
```

```python
#Lets create pipelines now 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
num_attribs = ["temp", "rain_1h","snow_1h","clouds_all",]
cat_attribs = ["holiday", "date_time"]
cat_attribs2 = ["weather_main"]
full_pipeline = ColumnTransformer([
    ("cat", CategoricalDealer(), cat_attribs),
    ("num", StandardScaler(), num_attribs),
    
    ("cat2", OneHotEncoder(), cat_attribs2),
])

metro_prepared = full_pipeline.fit_transform(metro)
```

```python
#Now lets try some models

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(metro_prepared, metro_labels)

some_data = metro.iloc[:5]
some_labels = metro_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print("Predictions: ", lin_reg.predict(some_data_prepared))
print("Labels: ", list(some_labels))
print(some_data.iloc[2]) #this one is outsider and it make predictions worse
#not so good :(
```

```python
#lets calculate RMSE
from sklearn.metrics import mean_squared_error
metro_predictions = lin_reg.predict(metro_prepared)
lin_mse = mean_squared_error(metro_labels, metro_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse #not so good We are underfitting
#definitely too big error
```

```python
#lets try RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(metro_prepared, metro_labels)

metro_predictions = forest_reg.predict(metro_prepared)
forest_mse = mean_squared_error(metro_labels, metro_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
#nice this one looks very promising 
#Ill cross Validate it later
```
